### Image ROI(Region of Image)

이미지의 일부분을 잘라낼 수 있다.

partial_img = img[y1:y2, x1:x2]

응용하면 복제도 가능하다.

```python
img = cv2.imread("../Data/desert.jpg")
b, g, r = cv2.split(img)   # img파일을 b,g,r로 분리
img2 = cv2.merge([r,g,b]) # b, r을 바꿔서 Merge

plt.imshow(img2)
plt.xticks([]) # x축 눈금
plt.yticks([]) # y축 눈금
plt.show()
```

```python
img3 = img
man = img[130:158, 135:146]
img3[126:154, 115:126] = man
b, g, r = cv2.split(img3)   # img파일을 b,g,r로 분리
img3 = cv2.merge([r,g,b]) # b, r을 바꿔서 Merge

plt.imshow(img3)
plt.xticks([]) # x축 눈금
plt.yticks([]) # y축 눈금
plt.show()
```
